In this quick analysis, I'm gonna take a look at how the economic and population indicator have changed over time.

## Libraries Import

```python
__author__ = 'geomars'

import os.path

# Import packages
import pandas as pd
import numpy as np
import datetime as dt

# Matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
# use the inline backend to generate the plots within the browser
%matplotlib inline 

# Seaborn
import seaborn as sns
sns.set(color_codes=True)
```

## Data Import and Extraction

```python
# Import the dataset
raw = pd.read_csv("../datasets/data_indonesia.csv", sep=',', \
                  dtype = None, error_bad_lines=False, \
                  encoding='utf-8', keep_default_na=False)                  
                  
# Extract the data into pandas dataframe
df = pd.DataFrame(raw)

# Drop rows with missing values and drop duplicate (optional)
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
x = df['year']
plt.plot(x, df['poverty_pct'], label="Poverty Percentage", c='blue', linewidth=2)
plt.plot(x, df['old_population_pct'], label="Older Population", c='black', linewidth=2)  
plt.plot(x, df['population_growth'], label="Population Growth", c='green', linewidth=2)
plt.plot(x, df['gdp'], label="GDP", c='red', linewidth=2)

plt.grid(color='lightblue',linestyle='-.', linewidth=0.7)
plt.legend()
plt.show()
```
